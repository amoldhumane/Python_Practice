```python
df=pd.read_csv('Social_Network_Ads - Social_Network_Ads.csv')
```

```python
df.head()
```

```python
X=df.drop('Purchased',axis=1)
```

```python
y=df['Purchased']
```

```python
# lets first normalise the values of X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar=MinMaxScaler()
```

```python
scalar.fit(X)
```

```python
X_scaled=scalar.transform(X)
```

```python
X_scaled
```

```python
y
```

```python
# Now we will plot all data point of x
##### where the Purchased value = 0 (will be plotted in red)
##### where the Purchased value = 1 (will be plotted in blue)
```

```python
X_Purch_Zero=X_scaled[y==0]   ## this will contain all values of X where product is not purchased
```

```python
X_Purch_Zero                  ## column1: Age and Column2: EstimatedSalary
```

```python
X_Purch_One=X_scaled[y==1]    ## this will contain all values of X where product is purchased
```

```python
X_Purch_One
```

```python
# Now lets plot the graph for product purchased and not purchased from the entire dataset (total 400 records)
```

```python
import matplotlib.pyplot as plt
# Lets seperate the cutomers who have not purchased the prduct and plot the graph with red marking (x axis-age and yaxis-estimated_salary)
column0=X_Purch_Zero[:,0]    #This will denote product not purchased
column1=X_Purch_Zero[:,1]
plt.scatter(column0,column1,color='red',marker='*')
column0=X_Purch_One[:,0]    #This will denote product is purchased
column1=X_Purch_One[:,1]
plt.scatter(column0,column1,color='blue',marker='+')
```

```python
# Now, lets spit the entire dataset into train and test subsets
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=.75,random_state=17)
```

```python
xtrain_scaled=scalar.transform(xtrain)
xtest_scaled=scalar.transform(xtest)
```

```python
# Now we will train the logistic regression model using training data
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain_scaled,ytrain)
```

```python
ypredicted=model.predict(xtest_scaled)      
```

```python
ypredicted
```

```python
model.score(xtest_scaled,ytest)    # score of model against the actual output
```

```python
xtest_scaled                       # it has two columns -> 1. Age and 2. EstimatedSalary
```

```python
# Now lets plot the scatterplot for xtest_scaled data for 'Product Purchased' and 'Product Not Purchased'
```

```python
xtest_scaled_prod_purchased=xtest_scaled[ypredicted==1]        #select the customers who have purchased the product
xtest_scaled_prod_not_purchased=xtest_scaled[ypredicted==0]    #select the customers who have not purchased the product


C0_PP=xtest_scaled_prod_purchased[:,0]    # Age of student
C1_PP=xtest_scaled_prod_purchased[:,1]    # EstimatesSalary of student

C0_PNP=xtest_scaled_prod_not_purchased[:,0]    # Age of student
C1_PNP=xtest_scaled_prod_not_purchased[:,1]    # EstimatesSalary of student
```

```python
plt.scatter(C0_PP,C1_PP,marker='*',color='blue')
plt.scatter(C0_PNP,C1_PNP,marker='*',color='red')
```
